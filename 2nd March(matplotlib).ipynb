Q2:

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("this is title")
plt.show()
```

```python
import numpy as np
fig = plt.figure()
gs = fig.add_gridspec(4, hspace = 1)
axs = gs.subplots(sharex=True, sharey=True)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
axs[0].plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
axs[1].plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
axs[2].plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
axs[3].plot(x,y)

for ax in axs:
    ax.label_outer()

```

Q3: 
Barplot is used visualize categorical vs numerical data

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit, color = 'lightblue')
plt.xlabel("Company's name")
plt.ylabel("profit")
plt.show()
```
